When calculating the p-value, we want to imagine a world where the null hypothesis is true, because the p-value helps us assess how unusual the data is under that hypothesis. For example, if we want to test whether our dog Fido has been known to get into the trash, the null hypothesis assumes that Fido is innocent. By imagining a world in which Fido has not touched the trash, we can ask: If Fido is innocent, what is the probability that we will see the same dirty situation in the kitchen? This probability is called the p-value and is used to measure how strange or rare the evidence we observe is in a world where the null hypothesis is true. If the p-value is very small, it suggests that this innocent-world scenario is unlikely, leading us to reject the null hypothesis and consider Fido guilty. Essentially, we determine whether the observed data is normal or surprising in this situation by imagining the null hypothesis to be true, which helps us make a judgment about the hypothesis.

## 5. Güntürkün (2003) recorded how kissing couples tilt their heads. 80 out of 124 couples, or 64.5% tilted their heads to the right. Simulate a p-value using a "50/50 coin-flipping" model for the assumption of the null hypothesis  𝐻0  that the population of humans don't have left or right head tilt tendencies when kissing, and use the table below to determine the level of evidence we have against  𝐻0 .

To simulate a p-value using a "50/50 coin-flipping" model for the given scenario, we can follow these steps:

1. Set Up the Null Hypothesis (H₀): We assume that there is no preference for head tilting when kissing. This means the probability of tilting left or right is equal (50% each).


2. Data Summary:
- Total couples: 124
- Couples tilting to the right: 80
- Proportion tilting right: 64.5%

3. Simulation Process:
- We will simulate the outcomes of 124 coin flips (representing the 124 couples), where each flip results in a head tilt to the right (1) or left (0) with equal probability (50%).
- We'll repeat this simulation a number of times (e.g., 10,000 times) to see how often we get a result of 80 or more couples tilting to the right.

```python
import numpy as np

# Parameters
total_couples = 124
right_tilts_observed = 80
num_simulations = 10000

# Simulate the null hypothesis (50/50)
simulated_right_tilts = np.random.binomial(total_couples, 0.5, num_simulations)

# Calculate the p-value
p_value = np.mean(simulated_right_tilts >= right_tilts_observed)
p_value
```

Interpretation:
This p-value indicates that, under the null hypothesis (H₀) of no preference for head tilt when kissing (i.e., a 50/50 chance of tilting left or right), the probability of observing 80 or more couples tilting their heads to the right is very low (0.08%).

Therefore, based on the table of level of evidence, we've noticed that this is a very strong evidence against the null hypothesis

A smaller p-value does not definitively prove that the null hypothesis is false; rather, it indicates that the observed data is unlikely under the assumption that the null hypothesis is true. In the case of Fido, we cannot conclusively determine his guilt or innocence based solely on a p-value. Ultimately, hypothesis testing provides evidence that can support or challenge a hypothesis, but it cannot deliver definitive proof of either outcome.

Problem Introduction

The tea experiment, originally conducted by Ronald Fisher and Dr. Muriel Bristol, aimed to assess whether the order of pouring tea and milk affects the perceived taste. In this adaptation, we analyze a sample of STA130 students to determine if they can correctly identify the order of pouring in a similar setup. This experiment provides insights into sensory perception and statistical hypothesis testing.

Relationship Between This Experiment and the Original with Fisher and Bristol

In Fisher and Bristol's original experiment, Dr. Bristol was able to identify the order of milk and tea in all eight cups presented to her. In contrast, we surveyed a random sample of 80 STA130 students, where 49 students correctly identified the pouring order. While the original study focused on a specific individual's ability to discern the taste difference, our study involves a larger group and seeks to understand whether their success rate significantly exceeds chance.

Statements of the Null Hypothesis and Alternative Hypothesis

Null Hypothesis (H₀)
The null hypothesis states that there is no difference in the ability to identify the pouring order. Formally, we can represent this as: H0 : p = 0.5 where p is the proportion of students who can correctly identify the order of pouring.

Informal Interpretive Statement
In simpler terms, we are saying that STA130 students have no better chance of correctly identifying the order of pouring tea and milk than if they were just guessing randomly, which would give them a 50% success rate.

Alternative Hypothesis (H₁)
The alternative hypothesis posits that students can identify the pouring order better than chance:

HA: p > 0.5

Quantitative Analysis

To test our hypotheses, we will conduct a hypothesis test for a proportion. Our observed test statistic is based on the proportion of students who correctly identified the order: p^ =49/80=0.6125

```python
import numpy as np
import scipy.stats as stats

# Parameters
n = 80  # Sample size
x = 49  # Number of successes
p_0 = 0.5  # Null hypothesis proportion

# Observed proportion
p_hat = x / n

# Standard error
se = np.sqrt(p_0 * (1 - p_0) / n)

# Test statistic (z-score)
z = (p_hat - p_0) / se

# P-value for one-tailed test
p_value = 1 - stats.norm.cdf(z)

# Results
z, p_value
```

```python
import matplotlib.pyplot as plt

# Bar plot for observed vs expected proportions
labels = ['Observed Proportion', 'Expected Proportion']
proportions = [p_hat, p_0]

plt.bar(labels, proportions, color=['blue', 'orange'])
plt.ylabel('Proportion')
plt.title('Observed vs Expected Proportions')
plt.ylim(0, 1)
plt.axhline(y=p_0, color='gray', linestyle='--', label='Null Hypothesis Proportion (0.5)')
plt.legend()
plt.show()
```

Findings and Discussion

Upon executing the code, we obtain the following results:

- Calculated z-score: The z-score indicates how many standard deviations our observed proportion is away from the null hypothesis proportion.
- P-value: The p-value quantifies the evidence against the null hypothesis. If the p-value is less than our significance level (commonly 0.05), we reject the null hypothesis.
For instance, if our p-value is 0.1, we would fail to reject H0, indicating that the difference in proportions could be due to random chance. Conversely, a p-value of 0.03 would suggest a significant ability of students to identify the order.

Conclusion Regarding the Null Hypothesis

Based on the calculated p-value and z-score, we draw a conclusion regarding our null hypothesis. If we reject H0, it suggests that STA130 students have a significant ability to discern the order of pouring tea and milk, indicating that sensory perception in this context may differ from random guessing.
