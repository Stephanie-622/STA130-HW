   - Visualization: I will choose box plots as it can be used to display salary distributions across different degree levels. Also I will use bar charts with average or median salaries per degree level may also illustrate general trends effectively.

	b. Visualizations (Part 2)

	c. Hypothesis Testing

Analysis 2: 

1. Research Question

    - Question: “How does the city of residence impact life satisfaction?”
    - Propose: This analysis aims to understand whether individuals living in different cities report varying levels of life satisfaction. Insights from this analysis can help identify how urban environments influence well-being and may guide city planners and policymakers in improving quality of life.

2. Variables and Plan for Exploration:

	- Independent Variable: City of residence (e.g., City A, City B, City C)
	- Dependent Variable: Life satisfaction score (a numerical measure of individuals’ self-reported life satisfaction)

    Why Choose These Variables?: City of residence can influence various factors that contribute to life satisfaction, such as economic opportunities, social services, recreational facilities, and overall quality of life. Understanding this relationship can shed light on the socio-economic impacts of urban living.

    Visualization:
    - Box Plots: To visualize the distribution of life satisfaction scores across different cities. This will allow for an easy comparison of medians, quartiles, and the presence of outliers among the cities.
    - Bar Charts: To display average life satisfaction scores for each city. This can effectively illustrate general trends and highlight any significant differences in life satisfaction levels between cities.

3. Analysis Plan:

    a. Summary Statistics:
    - Calculate Average Life Satisfaction: For each city to get a sense of general life satisfaction trends.
    - Calculate Median Life Satisfaction: To provide a clearer picture of typical life satisfaction scores, particularly if the data is skewed.

    b. Visualizations (Part 2):
	- Use the previously mentioned box plots and bar charts to visually summarize the data.

    c. Hypothesis Testing:
    
	- Null Hypothesis (H₀): There is no difference in average life satisfaction scores among individuals residing in different cities.
	- Alternative Hypothesis (H₁): There is a difference in average life satisfaction scores between at least two cities.

    Interpretation:
    - If the p-value is less than 0.05, we can conclude that there is a statistically significant difference in life satisfaction across cities.
    - If the p-value is greater than or equal to 0.05, we conclude that we do not have enough evidence to say that city of residence impacts life satisfaction.


4. Hypothesis:Individuals residing in cities with better amenities, social services, and economic opportunities are likely to report higher life satisfaction on average compared to those living in less favorable urban environments.

    - Relevance: Confirming a relationship between city residence and life satisfaction can provide valuable insights for policymakers, urban planners, and community organizations. It may encourage targeted initiatives to enhance the quality of life in specific areas, thus benefiting residents and potentially leading to greater overall societal well-being.


Analysis 3: 

1. Research Question

    Question: “How does age impact sleep difficulty?”
    Propose: This analysis aims to understand whether individuals of different age groups experience varying levels of sleep difficulty. Insights from this analysis can help identify age-related trends in sleep patterns and inform healthcare professionals and policymakers about the need for targeted interventions to improve sleep quality across different demographics.

2. Variables and Plan for Exploration

	- Independent Variable: Age (measured in years or categorized into age groups, e.g., 18-29, 30-44, 45-59, 60+)
	- Dependent Variable: Sleep difficulty score (a numerical measure based on self-reported difficulty sleeping, which may include factors such as trouble falling asleep, waking up frequently, or feeling rested upon waking)

    Why Choose These Variables?: Age is a critical factor influencing sleep patterns, with various studies indicating that sleep difficulty can increase with age due to physiological, psychological, and lifestyle changes. Examining this relationship can shed light on the necessity for age-specific interventions to enhance sleep quality.

    Visualization:
    - Box Plots: To visualize the distribution of sleep difficulty scores across different age groups. This will facilitate comparisons of medians, quartiles, and outliers among the age categories.
    - Charts: To display average sleep difficulty scores for each age group, effectively illustrating general trends and highlighting significant differences in sleep experiences across ages.


3. Analysis Plan

    a. Summary Statistics:
    - Calculate Average Sleep Difficulty: For each age group to understand general trends in sleep difficulty by age.
    - Calculate Median Sleep Difficulty: To provide a clearer view of typical sleep difficulties, particularly if the data is skewed.

    b. Visualizations (Part 2):
    - Utilize the mentioned box plots and bar charts to visually summarize and interpret the data.

    c. Hypothesis Testing:

    - Hypotheses:
        - Null Hypothesis (H₀): There is no difference in average sleep difficulty scores among individuals of different age groups.
        - Alternative Hypothesis (H₁): There is a difference in average sleep difficulty scores between at least two age groups.

    - Interpretation:
        - If the p-value is less than 0.05, we can conclude that there is a statistically significant difference in sleep difficulty across age groups.
        - If the p-value is greater than or equal to 0.05, we conclude that we do not have enough evidence to say that age impacts sleep difficulty.

4. Hypothsis: Older adults are likely to report higher levels of sleep difficulty on average compared to younger individuals, reflecting the natural changes in sleep patterns associated with aging.

    - Relevance: Confirming a relationship between age and sleep difficulty can provide critical insights for healthcare providers, enabling them to tailor interventions aimed at improving sleep quality in older populations. This analysis could lead to better health outcomes and enhance the overall well-being of individuals experiencing sleep difficulties as they age.
